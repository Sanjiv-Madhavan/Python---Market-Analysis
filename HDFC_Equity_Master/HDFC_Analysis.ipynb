```python
dataset_eqs = dataset.iloc[EqsData + 1:IncomeData]
dataset_eqs.columns = ['HOUSING DEVELOPMENT FIN. CORP. LTD.', 'No. of Months Year Ending', '12-Mar-15', '12-Mar-16', '12-Mar-17', '12-Mar-18', '12-Mar-19']
dataset_eqs = dataset_eqs.fillna("0")
# dataset_eqs.iloc[11, 2:7]
dataset_eqs.set_index("HOUSING DEVELOPMENT FIN. CORP. LTD.", inplace = True)
dataset_eqs
```

```python
cols = ['12-Mar-15', '12-Mar-16', '12-Mar-17', '12-Mar-18', '12-Mar-19']
dataset_eqs[cols] = dataset_eqs[cols].apply(lambda x: pd.to_numeric(x.astype(str)
                                                   .str.replace(',',''), errors='coerce'))
dataset_eqs
```

```python
m_rows = dataset_eqs[dataset_eqs['No. of Months Year Ending'].str.contains("Rs m") | dataset_eqs['No. of Months Year Ending'].str.contains("m")].index.values
m_rows

th_rows = dataset_eqs[dataset_eqs['No. of Months Year Ending'].str.contains("Rs Th")].index.values
th_rows

# cols = ['12-Mar-15', '12-Mar-16', '12-Mar-17', '12-Mar-18', '12-Mar-19']
# # dataset_cashflow[cols] = dataset_cashflow[cols].apply(lambda x: pd.to_numeric(x * 1000000 , errors='coerce'))
dataset_eqs = dataset_eqs.drop(['No. of Months Year Ending'], axis = 1)

dataset_eqs = dataset_eqs.apply(lambda x: x * 1000000 if x.name in m_rows else x, axis=1)
dataset_eqs = dataset_eqs.apply(lambda x: x * 1000 if x.name in th_rows else x, axis=1)
dataset_eqs = dataset_eqs.T
dataset_eqs
```
